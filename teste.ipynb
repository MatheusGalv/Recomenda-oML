```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_excel("data/Rating.xlsx")
```

```python
df = pd.DataFrame(data)

df_pivotada = df.pivot_table(index='Nome', columns='Filme', values='Rating', fill_value=0)

df_pivotada.head()
```

```python
matriz_similaridade = cosine_similarity(df_pivotada)
df_similaridade = pd.DataFrame(matriz_similaridade, index=df_pivotada.index, columns=df_pivotada.index)
df_similaridade.head(8)
```
